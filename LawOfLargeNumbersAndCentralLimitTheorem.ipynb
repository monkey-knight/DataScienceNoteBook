## 1.1 大数定律

### 大数定律 Python 代码模拟

```python
from scipy.stats import bernoulli

p = 0.5  # 假定伯努利分布 p = 0.5
real_mean = bernoulli.mean(p)  # 则伯努利分布的平均值为 0.5
print("平均值真实值：", real_mean)
```

```python
import numpy as np

sample_num_list = np.array([i + 1 for i in range(20000)])  # 产生一个列表，列表中的元素表示采样的样本数量，样本数量依次增加
print("samples: ", sample_num_list)
```

```python
means = []
for sample_num in sample_num_list:
    samples = bernoulli.rvs(p, size=sample_num)  # 依据样本数量，生成随机样本
    means.append(np.mean(samples))  # 计算本次采样的样本均值
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.plot(sample_num_list, np.array(means))
plt.xlabel("number of samples")
plt.ylabel("mean value of samples")
plt.show()
```

由上述实验结果，我们可以发现：**当采样的样本数量越大，其样本均值就越接近于真实的均值**。
